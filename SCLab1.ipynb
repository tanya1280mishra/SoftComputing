```python
print("Probabilistic Sum: A+^B")
valA=[0.23,0.25,0.17,0.67,0.678]
valB=[0.23,0.67,0.17,0.89,0.789]
val=valA+valB

def cal_probability(val,valA):
    count=0
    prob=0
    for it in val:
        for vala in valA:
            if vala==it:
                count=count+1

    prob=count/len(val)
    return prob

def inter_probability(val,valA,valB):
    count=0
    for it in val:
        for vala in valA:
            for valb in valB:
                if vala==valb and valb==it:
                    count=count+1
                    
    prob=count/len(val)
    return prob
    
UA=cal_probability(val,valB)
UB=cal_probability(val,valA)
UAUB=inter_probability(val,valA,valB)
prob=cal_probability(val,valB)+cal_probability(val,valA)-inter_probability(val,valA,valB)
print(UA)
print(UB)
print(UAUB)
print(prob)
```

```python
print("Euclidean distance between points of valA and valB is:")
import math
def euclid(valA,valB):
    # Calculate the Euclidean distance  
    squared_differences = [(valA[i] - valB[i]) ** 2 for i in range(len(valA))]  
    distance = math.sqrt(sum(squared_differences))  

    return distance  


print(euclid(valA,valB))
```

```python
print("Hamilton Distance")
#Hamming Distance Formula
# The Hamming distance between two strings of equal length is defined as the number of positions at which
# the corresponding symbols are different.
def hamming_distance(valA, valB):  
    # Check if the lengths of the two lists are equal  
    if len(valA) != len(valB):  
        raise ValueError("Both lists must have the same length.")  

    dis = 0  # Initialize distance counter  
    for vala, valb in zip(valA, valB):  
        if vala != valb:  
            dis += 1  # Increment distance for each differing position  

    return dis  

distance = hamming_distance(valA, valB)  
index = len(valA)  # Length of the input lists (same for both)  

print("Hamming Distance:", distance)  
print("Length of the input lists:", index)
plt.figure(figsize=(10,5))
plt.plot(index,distance,marker='o',color='b',label='data line')

plt.title('Hamilton Distance graph')
plt.xlabel('Distance')
plt.ylabel('member')
plt.xticks(mem)
plt.grid(True)
plt.legend()
plt.show()
```

```python
print("Demorgan LawI")
print("~(AuB)=(~An~B)")
def law1(valA,valB):
    law=[]
    for vala,valb in zip(valA,valB):
        law.append(min(1-vala,1-valb))
        return law
first=[]
first=law1(valA,valB)
for i in first:
    print(i)


print("Demorgan LawII")
print("~(AnB)=(~Au~B)")
def law2(valA,valB):
    law=[]
    for vala,valb in zip(valA,valB):
        law.append(max(1-vala,1-valb))
        return law
sec=[]
sec=law2(valA,valB)
for j in sec:
    print(j)


print("Law of contradiction")
def contra(valA):
    con=[]
    for it in valA:
        con.append(min(it,1-it))
        return con
CA=[]
CB=[]
CA=contra(valA)
for i in CA:
    print(i)
CB=contra(valB)
for j in CB:
    print(j)


print("Law of Excluded Middle")
def Exclu(valA):
    ex=[]
    for it in valA:
        ex.append(max(it,1-it))
        return ex
exa=[]
exb=[]
exa=Exclu(valA)
for i in exa:
    print(i)
exb=Exclu(valB)
for i in exb:
    print(i)
```

```python

print("Membership functions")
import numpy as np  
import matplotlib.pyplot as plt  

def trapezoidal_membership(x, a, b, c, d):  
    """  
    Trapezoidal membership function.  
    
    Parameters:  
    x : float or np.array  
        Input value(s).  
    a, b, c, d : float  
        Parameters defining the trapezoid.  
        
    Returns:  
    float or np.array  
        Membership value(s).  
    """  
    return np.clip((x - a) / (b - a), 0, 1) * (x <= c) + \
           np.clip((d - x) / (d - c), 0, 1) * (x > c)  

def sigmoid_membership(x, a, b):  
    """  
    Sigmoid membership function.  
    
    Parameters:  
    x : float or np.array  
        Input value(s).  
    a : float  
        The x-value of the sigmoid's midpoint.  
    b : float  
        The steepness of the curve.  
        
    Returns:  
    float or np.array  
        Membership value(s).  
    """  
    return 1 / (1 + np.exp(-b * (x - a)))  

# Example usage  
x = np.linspace(-10, 10, 100)  

# Trapezoidal membership function parameters  
a, b, c, d = -5, -2, 2, 5  
trapezoidal_values = trapezoidal_membership(x, a, b, c, d)  

# Sigmoid membership function parameters  
sigmoid_a, sigmoid_b = 0, 1  
sigmoid_values = sigmoid_membership(x, sigmoid_a, sigmoid_b)  

# Plotting the membership functions  
plt.figure(figsize=(12, 6))  

# Trapezoidal Membership Function  
plt.subplot(1, 2, 1)  
plt.plot(x, trapezoidal_values, label='Trapezoidal Membership', color='blue')  
plt.title('Trapezoidal Membership Function')  
plt.xlabel('x')  
plt.ylabel('Membership Degree')  
plt.grid()  
plt.axhline(0, color='black', lw=0.5, ls='--')  
plt.axvline(0, color='black', lw=0.5, ls='--')  
plt.legend()  

# Sigmoid Membership Function  
plt.subplot(1, 2, 2)  
plt.plot(x, sigmoid_values, label='Sigmoid Membership', color='green')  
plt.title('Sigmoid Membership Function')  
plt.xlabel('x')  
plt.ylabel('Membership Degree')  
plt.grid()  
plt.axhline(0, color='black', lw=0.5, ls='--')  
plt.axvline(0, color='black', lw=0.5, ls='--')  
plt.legend()  

plt.tight_layout()  
plt.show()
```

```python
import numpy as np  
import matplotlib.pyplot as plt  

def generalized_bell_membership(x, a, b, c):  
    """  
    Generalized Bell membership function.  
    
    Parameters:  
    x : float or np.array  
        Input value(s).  
    a : float  
        The width of the bell.  
    b : float  
        The position of the center of the bell.  
    c : float  
        The slope of the bell.  
        
    Returns:  
    float or np.array  
        Membership value(s).  
    """  
    return 1 / (1 + np.abs((x - b) / a) ** (2 * c))  

def gaussian_membership(x, mean, sigma):  
    """  
    Gaussian membership function.  
    
    Parameters:  
    x : float or np.array  
        Input value(s).  
    mean : float  
        The mean (center) of the Gaussian.  
    sigma : float  
        The standard deviation (width) of the Gaussian.  
        
    Returns:  
    float or np.array  
        Membership value(s).  
    """  
    return np.exp(-0.5 * ((x - mean) / sigma) ** 2)  

def triangular_membership(x, a, b, c):  
    """  
    Triangular membership function.  
    
    Parameters:  
    x : float or np.array  
        Input value(s).  
    a, b, c : float  
        Parameters defining the triangle (a: left vertex, b: peak, c: right vertex).  
        
    Returns:  
    float or np.array  
        Membership value(s).  
    """  
    return np.clip((x - a) / (b - a), 0, 1) * (x <= b) + \
           np.clip((c - x) / (c - b), 0, 1) * (x > b)  

# Example usage  
x = np.linspace(-10, 10, 100)  

# Generalized Bell membership function parameters  
bell_a, bell_b, bell_c = 2, 0, 2  
bell_values = generalized_bell_membership(x, bell_a, bell_b, bell_c)  

# Gaussian membership function parameters  
gaussian_mean, gaussian_sigma = 0, 2  
gaussian_values = gaussian_membership(x, gaussian_mean, gaussian_sigma)  

# Triangular membership function parameters  
triangular_a, triangular_b, triangular_c = -5, 0, 5  
triangular_values = triangular_membership(x, triangular_a, triangular_b, triangular_c)  

# Plotting the membership functions  
plt.figure(figsize=(12, 6))  

# Generalized Bell Membership Function  
plt.subplot(1, 3, 1)  
plt.plot(x, bell_values, label='Generalized Bell Membership', color='blue')  
plt.title('Generalized Bell Membership Function')  
plt.xlabel('x')  
plt.ylabel('Membership Degree')  
plt.grid()  
plt.axhline(0, color='black', lw=0.5, ls='--')  
plt.axvline(0, color='black', lw=0.5, ls='--')  
plt.legend()  

# Gaussian Membership Function  
plt.subplot(1, 3, 2)  
plt.plot(x, gaussian_values, label='Gaussian Membership', color='green')  
plt.title('Gaussian Membership Function')  
plt.xlabel('x')  
plt.ylabel('Membership Degree')  
plt.grid()  
plt.axhline(0, color='black', lw=0.5, ls='--')  
plt.axvline(0, color='black', lw=0.5, ls='--')  
plt.legend()  

# Triangular Membership Function  
plt.subplot(1, 3, 3)  
plt.plot(x, triangular_values, label='Triangular Membership', color='red')  
plt.title('Triangular Membership Function')  
plt.xlabel('x')  
plt.ylabel('Membership Degree')  
plt.grid()  
plt.axhline(0, color='black', lw=0.5, ls='--')  
plt.axvline(0, color='black', lw=0.5, ls='--')  
plt.legend()  

plt.tight_layout()  
plt.show()
```
